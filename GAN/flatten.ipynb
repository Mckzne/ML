```python
import os
import shutil

original_root = '/kaggle/input/task1-brain/brain'
flattened_root = '/kaggle/working/brain_flat'

os.makedirs(flattened_root, exist_ok=True)

for patient_id in os.listdir(original_root):
    patient_path = os.path.join(original_root, patient_id)
    new_patient_path = os.path.join(flattened_root, patient_id)
    os.makedirs(new_patient_path, exist_ok=True)

    for folder in ['mr.nii', 'ct.nii', 'mask.nii']:
        folder_path = os.path.join(patient_path, folder)
        file_path = os.path.join(folder_path, folder)  # actual file inside the same-named folder
        if os.path.exists(file_path):
            dest_path = os.path.join(new_patient_path, folder.replace('.nii', '') + '.nii')
            shutil.copy(file_path, dest_path)

print(" Dataset flattened to /kaggle/working/brain_flat")
```
